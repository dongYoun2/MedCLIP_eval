```python
df_sample = df_merged.sample(n=500, random_state=42).reset_index(drop=True)
print(f"Sampled rows: {len(df_sample)}")
```

```python
import os, shutil
```

```python
OUT_DIR = "./indiana_sampled_500_images/images_normalized"
os.makedirs(OUT_DIR, exist_ok=True)
```

```python
copied, missing = 0, 0
dst_paths = []
for fn in df_sample["filename"]:
    src = os.path.join(IMG_PATH, str(fn))
    dst = os.path.join(OUT_DIR, os.path.basename(str(fn)))
    if os.path.exists(src):
        shutil.copy2(src, dst)  # preserves metadata
        dst_paths.append(dst)
        copied += 1
    else:
        dst_paths.append(None)
        missing += 1

df_sample["copied_path"] = dst_paths
print(f"Copied: {copied}, Missing: {missing}, Output dir: {OUT_DIR}")
```

```python
df_sample.to_csv("indiana_sampled_500_images.csv", index=False)
```
